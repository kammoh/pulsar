```python
import sys
import numpy as np
import pandas as pd
# import seaborn as sns
import matplotlib.pyplot as plt
import plotly.offline as plotly
import plotly.graph_objs as go


plotly.init_notebook_mode(connected=True)
```

```python
# sns.set(style="darkgrid")

df = pd.read_csv("1.csv", header=None, dtype='i8', sep='[^\d]*', engine='python', skiprows=1, usecols=[1,2])



xbins=dict(
    start=1,
    end=200,
    size=5
)

trace1 = go.Bar(
    y=df.iloc[:,0],
    opacity=0.75, 
)
trace2 = go.Bar(
    y=df.iloc[:,1],
    opacity=0.75
)

data = [trace1, trace2]
layout = go.Layout(barmode='overlay')
fig = go.Figure(data=data, layout=layout)

plotlyiplot(fig, filename='overlaid histogram')
```
